#### Filling in Missing Values

```python
# get a small subset of the data
nfl_data_subset = nfl_data.loc[:, 'EPA':'Season'].head()
nfl_data_subset.head()
```

```python
# replace NAs with 0
nfl_data_subset.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
nfl_data_subset.fillna(method='bfill', axis=0).fillna(0)
```

```python
for colname in nfl_data.select_dtypes("object"):
    nfl_data[colname], _ = nfl_data[colname].factorize()

nfl_data_num = nfl_data.dtypes == int
nfl_data_num

nfl_data_num = nfl_data.drop(nfl_data.select_dtypes("object"), axis=1)
nfl_data_num.head()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

nfl_data_prep = imputer.fit(nfl_data_num)
nfl_data_prep
```
