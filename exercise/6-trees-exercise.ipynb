```python
from collections import deque

class BTNode:
    def __init__(self, data):
        self.right = None
        self.left = None
        self.data = data
        
    def display(self):
        return print(self.data, end=' ')
        
class BinaryTree:
    def __init__(self):
        self.root = None
        
    def insert(self, item):   # Level order (breadth first) insertion
        new_node = BTNode(item)
        if self.root == None:
            self.root = new_node
        
        queue = deque([self.root])
        while queue:
            ptr = queue.popleft()
            
            if not ptr.left:
                ptr.left = new_node
                return
            queue.append(ptr.left)
            
            if not ptr.right:
                ptr.right = new_node
                return
            queue.append(ptr.right)
            
    def delete(self, item):
        if not self.root:
            return "Tree is empty"
        
        if (self.root.data == item) and not (self.root.left or self.root.right):
            self.root = None
            return
        
        queue = deque([self.root])
        node_to_delete = None
        last_node = None
        
        while queue:
            last_node = queue.popleft()
            if last_node.data == item:
                node_to_delete = last_node
                
            if last_node.left:
                queue.append(last_node.left)
            if last_node.right:
                queue.append(last_node.right)
                
        if node_to_delete:
            node_to_delete.data = last_node.data
            self._delete_deepest(last_node)
        else:
            return "Value not found"
        
    def _delete_deepest(self, deepest_node):
        queue = deque([self.root])
        while queue:
            ptr = queue.popleft()
            
            if ptr.left:
                if ptr.left == deepest_node:
                    ptr.left = None
                    return
                queue.append(ptr.left)
                
                if ptr.right == deepest_node:
                    ptr.right = None
                    return
                queue.append(ptr.right)
                
    def level_order(self):
        if not self.root:
            return "Tree is empty"
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            print(node.data, end=' ')
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
```

```python
class BSTNode:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    def left_most(self):
        return self.left
    
    def right_most(self):
        return self.right
    
    def set_data(self, val):
        if self.data == None:
            self.data = val
            
    def display(self):
        print(self.data, end="")
        
        
class BinarySearchTree:
    def __init__(self, data1):
        self.root = BSTNode(data1)
        self.level = 0
        
    def preorder(self):
        if self.root == None:
            return []
        
        stack = [self.root]
        trav = []
        
        while stack:
            ptr = stack.pop()
            trav.append(ptr.data)
            
            if ptr:
                if ptr.left == None and ptr.right == None:
                    print("Left Node:", ptr.data)
                    continue
                if ptr.right != None:
                    stack.append(ptr.right)
                if ptr.left != None:
                    stack.append(ptr.left)
                    
        return trav
                        
    def inorder(self, root):
        if root == None:
            return []
        
        stack = []
        trav = []
        ptr = root
        
        while stack or ptr:
            if ptr:
                stack.append(ptr)
                ptr = ptr.left
            else:
                ptr = stack.pop()
                trav.append(ptr.data)
                ptr = ptr.right
                    
    def postorder(self, root):
        if root == None:
            return
        
        stack = []
        trav = []
        ptr = root
        last_visited = None
        
        while stack or ptr:
            if ptr:
                stack.append(ptr)
                ptr = ptr.left
            else:
                peek = stack[-1]
                if peek.right and last_visited != peek.right:
                    ptr = peek.right
                else:
                    last_visited = stack.pop()
                    trav.append(ptr.data)
                    
        return trav
                
    def insert_right(self, val):
        ptr = self.root
        if ptr.right == None:
            ptr.right = BSTNode(val)
        else:
            temp = BSTNode(val)
            temp.right = ptr.right
            ptr.right = temp
            
    def insert_left(self, val):
        ptr = self.root
        if ptr.left == None:
            ptr.left = BSTNode(val)
        else:
            temp = BSTNode(val)
            temp.left = ptr.left
            ptr.left = temp
            
    def insert(self, root, key):
        if root is None:
            return BSTNode(key)
        if key < root.key:
            root.left = self.insert(root.left, key)
        elif key > root.key:
            root.right = self.insert(root.right, key)
            
    def search(self, root, key):
        if root is None or root.key == key:
            return root
        if key < root.key:
            return self.search(root.left, key)
        else:
            return self.search(root.right, key)
        
    def delete(self, root, key):
        if root is None:
            return root
        if key < root.key:
            root.left = self.delete(root.left, key)
        elif key > root.key:
            root.right = self.delete(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            
            successor = self.minimum_value(root.right)
            root.key = successor.key
            root.right = self.delete(root.right, successor.key)
            
        return root
            
    def minimum_value(self, node):
        ptr = node
        while ptr.left != None:
            ptr = ptr.left
            
        return ptr
            
    def insert_level_order(self, arr, root, i, n):
        if i < n:
            temp = BSTNode(arr[i])
            root = temp
            root.left = self.insertLevelOrder(arr, root.left, 2 * i + 1, n)
            root.right = self.insertLevelOrder(arr, root.right, 2 * i + 2, n)
        return root
    
    def delete_leaf_node(self, root, key):
        if root is None:
            return None
        if root.left is None and root.right is None and root.data == key:
            return None
        root.left = self.deleteLeafNode(root.left, key)
        root.right = self.deleteLeafNode(root.right, key)
        return root
```

```python
class AVLNode:
    def __init__(self, key, data = None):
        self.key = key
        self.right = None
        self.left = None
        self.height = 1
        self.data = data
        
    def display(self):
        print(self.key, end=" ")
        

class AVLTree:
    def __init__(self, data1):
        self.root = AVLNode(data1)
        
    def get_height(self, node):
        if not node:
            return 0
        return node.height if node else 0
    
    def update_height(self, node):
        if not node:
            return 0
        node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))
        return node.height
    
    def get_balance_factor(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)
    
    def rotate_right(self, y):
        x = y.left
        tree2 = x.right
        
        x.right = y
        y.left = tree2
        
        self.update_height(y)
        self.update_height(x)
        
        return x
    
    def rotate_left(self, x):
        y = x.left
        tree2 = y.right
        
        y.right = x
        x.left = tree2
        
        self.update_height(x)
        self.update_height(y)
        
        return y
```
