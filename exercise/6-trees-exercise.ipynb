```python
from collections import deque

class BTNode:
    def __init__(self, data):
        self.right = None
        self.left = None
        self.data = data
        
    def display(self):
        return print(self.data, end=' ')
        
class BinaryTree:
    def __init__(self):
        self.root = None
        
    def insert(self, item):
        new_node = BTNode(item)
        if self.root is None:
            self.root = new_node
            return  # ✅ MUST stop here
        
        queue = deque([self.root])
        
        while queue:
            node = queue.popleft()
            
            if not node.left:
                node.left = new_node
                return
            else:
                queue.append(node.left)
            
            if not node.right:
                node.right = new_node
                return
            else:
                queue.append(node.right)
            
    def search(self, item):
        if not self.root:
            return "Tree is empty"
        
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            
            if node.data == item:
                return True
            
            if node.left:
                queue.append(node.left)
    # def search(self, root, key):
    #     if not root:
    #         return False
    #     return root.data == key or self.search(root.left, key) or self.search(root.right, key)
            
    def delete(self, item):
        if not self.root:
            return "Tree is empty"
        
        if (self.root.data == item) and not (self.root.left or self.root.right):
            self.root = None
            return
        
        queue = deque([self.root])
        node_to_delete = None
        last_node = None
        
        while queue:
            last_node = queue.popleft()
            
            if last_node.data == item:
                node_to_delete = last_node
                
            if last_node.left:
                queue.append(last_node.left)
            if last_node.right:
                queue.append(last_node.right)
                
        if node_to_delete:
            node_to_delete.data = last_node.data
            self._delete_deepest(last_node)
        else:
            return "Value not found"
        
    def _delete_deepest(self, deepest_node):
        queue = deque([self.root])
        while queue:
            ptr = queue.popleft()
            
            if ptr is deepest_node:
                ptr = None
                return
            
            if ptr.left:
                if ptr.left is deepest_node:
                    ptr.left = None
                    return
                queue.append(ptr.left)
            
            if ptr.right:
                if ptr.right is deepest_node:
                    ptr.right = None
                    return
                queue.append(ptr.right)
                
    def level_order(self):
        if not self.root:
            return "Tree is empty"
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            print(node.data, end=' ')
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
    def inorder(self, node):
        if node:
            # left --> root --> right
            self.inorder(node.left)
            print(node.data, end=' ')
            self.inorder(node.right)
            
    def preorder(self, node):
        if node:
            # root --> left --> right
            print(node.data, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)
            
    def postorder(self, node):
        if node:
            # left --> right --> root
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=' ')
            
    def inorder_(self, root):
        if not root:
            return []
        return self.inorder_(root.left) + [root.data] + self.inorder_(root.right)

    def preorder_(self, root):
        if not root:
            return []
        return [root.data] + self.preorder_(root.left) + self.preorder_(root.right)

    def postorder_(self, root):
        if not root:
            return []
        return self.postorder_(root.left) + self.postorder_(root.right) + [root.data]
            
    def _preorder(self, root):
        return [] if not root else [root.data] + self._preorder(root.left) + self._preorder(root.right)

    def _inorder(self, root):
        return [] if not root else self._inorder(root.left) + [root.data] + self._inorder(root.right)

    def _postorder(self, root):
        return [] if not root else self._postorder(root.left) + self._postorder(root.right) + [root.data]


# Example Usage:
if __name__ == "__main__":
    bst = BinaryTree()
    values = [50, 30, 20, 40, 70, 60, 80]
    
    for val in values:
        bst.insert(val)
        
    print("Inorder traversal:", bst.inorder(bst.root))
    print("Inorder traversal:", bst.inorder_(bst.root))
    print("Inorder traversal:", bst._inorder(bst.root))
    
    print("\nDelete 20")
    bst.delete(20)
    print("Inorder traversal after deleting 20:", bst.inorder(bst.root))
    print("Inorder traversal after deleting 20:", bst.inorder_(bst.root))
    print("Inorder traversal after deleting 20:", bst._inorder(bst.root))
```

```python
class BSTNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # INSERT
    def insert(self, key):
        self.root = self._insert(self.root, key)

    def _insert(self, root, key):
        if root is None:
            return BSTNode(key)
        if key < root.data:
            root.left = self._insert(root.left, key)
        elif key > root.data:
            root.right = self._insert(root.right, key)
        return root

    # DELETE
    def delete(self, key):
        self.root = self._delete(self.root, key)

    def _delete(self, root, key):
        if root is None:
            return None

        if key < root.data:
            root.left = self._delete(root.left, key)
        elif key > root.data:
            root.right = self._delete(root.right, key)
        else:
            if root.left is None:
                return root.right
            if root.right is None:
                return root.left
            successor = self._min_value(root.right)
            root.data = successor.data
            root.right = self._delete(root.right, successor.data)

        return root

    def _min_value(self, node):
        while node.left:
            node = node.left
        return node

    # INORDER TRAVERSAL
    def inorder(self, root):
        return self.inorder(root.left) + [root.data] + self.inorder(root.right) if root else []


# ✅ TESTING
if __name__ == "__main__":
    bst = BinarySearchTree()
    for val in [50, 30, 20, 40, 70, 60, 80]:
        bst.insert(val)

    print("Inorder traversal:", bst.inorder(bst.root))

    print("\nDelete 20")
    bst.delete(20)
    print("Inorder traversal:", bst.inorder(bst.root))

    print("\nDelete 30")
    bst.delete(30)
    print("Inorder traversal:", bst.inorder(bst.root))

    print("\nDelete 50")
    bst.delete(50)
    print("Inorder traversal:", bst.inorder(bst.root))
```

```python
class AVLNode:
    def __init__(self, key, data=None):
        self.key = key
        self.right = None
        self.left = None
        self.height = 1
        self.data = key  # Use key for traversal output

class AVLTree:
    def __init__(self, data1):
        self.root = AVLNode(data1)

    def get_height(self, node):
        return node.height if node else 0

    def update_height(self, node):
        node.height = 1 + max(self.get_height(node.left), self.get_height(node.right))
        return node.height

    def get_balance_factor(self, node):
        return self.get_height(node.left) - self.get_height(node.right)

    # rotate right at y (y becomes right child of x)
    def rotate_right(self, y):
        x = y.left
        T2 = x.right
        x.right = y
        y.left = T2
        self.update_height(y)
        self.update_height(x)
        return x

    # rotate left at x (x becomes left child of y) l0
    def rotate_left(self, x):
        y = x.right
        T2 = y.left
        y.left = x
        x.right = T2
        self.update_height(x)
        self.update_height(y)
        return y
    
    def left_right_rotate(self, x):
        x.left = self.rotate_left(x.left)
        return self.rotate_right(x)
    
    def right_left_rotate(self, x):
        x.right = self.rotate_right(x.right)
        return self.rotate_left(x)

    def insert(self, root, key):
        if not root:
            return AVLNode(key)

        if key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        self.update_height(root)
        balance = self.get_balance_factor(root)

        if balance > 1 and key < root.left.key:
            return self.rotate_right(root)
        if balance < -1 and key > root.right.key:
            return self.rotate_left(root)
        if balance > 1 and key > root.left.key:
            root.left = self.rotate_left(root.left)
            return self.rotate_right(root)
        if balance < -1 and key < root.right.key:
            root.right = self.rotate_right(root.right)
            return self.rotate_left(root)

        return root

    def delete(self, root, key):
        if not root:
            return root

        if key < root.key:
            root.left = self.delete(root.left, key)
        elif key > root.key:
            root.right = self.delete(root.right, key)
        else:
            if not root.left:
                return root.right
            elif not root.right:
                return root.left

            temp = self.get_min_value_node(root.right)
            root.key = temp.key
            root.data = temp.data
            root.right = self.delete(root.right, temp.key)

        self.update_height(root)
        balance = self.get_balance_factor(root)

        if balance > 1 and self.get_balance_factor(root.left) >= 0:
            return self.rotate_right(root)
        if balance < -1 and self.get_balance_factor(root.right) <= 0:
            return self.rotate_left(root)
        if balance > 1 and self.get_balance_factor(root.left) < 0:
            root.left = self.rotate_left(root.left)
            return self.rotate_right(root)
        if balance < -1 and self.get_balance_factor(root.right) > 0:
            root.right = self.rotate_right(root.right)
            return self.rotate_left(root)

        return root

    def get_min_value_node(self, root):
        current = root
        while current.left:
            current = current.left
        return current

    # Traversal Returning Lists
    def L0_rotate(self, x):
        return self.rotate_left(x)

# ✅ TEST
if __name__ == "__main__":
    avl = AVLTree(10)
    keys = [20, 30, 40, 50, 25]

    for key in keys:
        avl.root = avl.insert(avl.root, key)

    print("Preorder:", avl.preorder(avl.root))
    print("Inorder:", avl.inorder(avl.root))
    print("Postorder:", avl.postorder(avl.root))

    avl.root = avl.delete(avl.root, 10)
    print("\nDelete 10")
    print("Preorder after delete:", avl.preorder(avl.root))
    print("Inorder after delete:", avl.inorder(avl.root))
    print("Postorder after delete:", avl.postorder(avl.root))
```
