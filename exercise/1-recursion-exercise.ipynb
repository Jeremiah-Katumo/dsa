```python
def tower_of_hanoi(num, source, dest, aux):
    if num == 1:
        print(f"Move disk 1 from {source} to {dest}")
    else:
        tower_of_hanoi(num - 1, source, aux, dest)  # Move n-1 disks from source to auxiliary
        print(f"Move disk {num} from {source} to {dest}")
        tower_of_hanoi(num - 1, aux, dest, source)  # Move n-1 disks from auxiliary to destination
        
```

```python
def generate_binary(n):
    if n == 0:
        return ""
    elif n == 1:
        return ['0','1']
    else:
        listt = []
        for i in generate_binary(n-1):
            listt.append('0'+i)
        for i in generate_binary(n-1):
            listt.append('1'+i)
    return listt

def generate_binary(n, prefix=''):
    if n == 0:
        return [prefix]
    else:
        return generate_binary(n-1, prefix + '0') + generate_binary(n-1, prefix + '1')
    
print(f"Binary Numbers Generated: {generate_binary(3)}")
```

```python
def generate_hex(n, prefix=""):
    hex_digits = "0123456789ABCDEF"

    # Base case
    if len(prefix) == n:
        print(prefix)
        return

    # Recursive case: try each hex digit
    for digit in hex_digits:
        return generate_hex(n, prefix + digit)
       
print(f"Hexadecimal Numbers Generated: {generate_hex(2)}") 


def generate_hex(n, prefix=''):
    if n == 0:
        return [prefix]
    else:
        hex_chars = '0123456789ABCDEF'
        result = []
        for char in hex_chars:
            result.extend(generate_hex(n-1, prefix + char))
        return result
    
print(f"Hexadecimal Numbers Generated: {generate_hex(2)}")
```

```python
def reverse_list(listt):
    '''Recursive Approach'''
    if len(listt) == 1:
        return listt
    else:
        return [listt[-1]] + reverse_list(listt[:-1])

listt = [4,5,6,7,8]
tic = time.time()
reverse_list(listt)
toc = time.time()
elapsed = toc - tic
print(f"Recursive Reversed List: {reverse_list(listt)}, Elapsed In: {elapsed}")

def reverse_list(listt):
    '''Iterative Approach'''
    n = len(listt)
    result = []
    if len(listt) == 1:
        result = listt
    
    for i in range(n-1, -1, -1):
        result.append(listt[i])
        
    return result

tic = time.time()
reverse_list(listt)
toc = time.time()
elapsed = toc - tic            
print(f"Iterative Reversed List: {reverse_list(listt)}, ELapsed In: {elapsed}")

def reverse_list_recur(arr, left=0, right=None):
    '''Recursive'''
    if right is None:
        right = len(arr) - 1
        
    if left >= right:
        return arr
    
    arr[left], arr[right] = arr[right], arr[left]
    return reverse_list_recur(arr, left + 1, right - 1)

tic = time.time()
reverse_list_recur(listt)
toc = time.time()
elapsed = toc - tic
print(f"Recursive Reversed Array: {reverse_list(listt)}, Elapsed In: {elapsed}")

def reverse_list_iter(arr):
    '''Iterative'''
    left = 0
    right = len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
        
    return arr

tic = time.time()
reverse_list_iter(listt)
toc = time.time()
elapsed = toc - tic
print(f"Iterative Reversed Array: {reverse_list_iter(listt)}, Elapsed In: {elapsed}")
```

```python
def reverse_number(num):
    '''Recursive'''
    if num < 10:
        return str(num)
    else:
        return str(num % 10) + reverse_number(num // 10)
    
num = 3456789
print(f"Recursive Reversed Number: {reverse_number(num)}")

def reverse_number_recur(n, rev=0):
    if n == 0:
        return rev
    
    digit = n % 10
    return reverse_number_recur(n // 10, rev * 10 + digit)

print(f"Recursive Reversed Number: {reverse_number_recur(num)}")


def reverse_number(num):
    """Iterative"""
    result = ""
    
    while num > 0:
        result += str(num % 10)
        num //= 10
        
    return result 

print(f"Iterative Reversed Number: {reverse_number(num)}")

def reverse_number_iter(n):
    rev = 0
    
    while n > 0:
        digit = n % 10
        rev = rev * 10 + digit
        n //= 10
        
    return rev

print(f"Iterative Reversed Number: {reverse_number_iter(num)}")
```

```python
def find_max(listt):
    '''Recursive'''
    if len(listt) == 1:
        return listt[0]
    else:
        rest = find_max(listt[1:])
        return listt[0] if listt[0] > rest else rest

listt = [89, 62, 78, 28, 89]
print(f"Recursive Maximum Number: {find_max(listt)}")

def find_max(arr, n):
    '''Iterative'''
    max1 = 0
    for i in range(1, n):
        if arr[i] > max1:
            max1 = arr[i]
    return max1

print(f"Iterative Maximum Number: {find_max(listt, 5)}")
```

```python
def find_min(arr):
    '''Recursive'''
    if len(arr) == 1:
        return arr[0]
    
    rest = find_min(arr[1:])
    return arr[0] if arr[0] < rest else rest

listt = [89, 62, 78, 28, 89]
print(f"Recursive Minimum Number: {find_min(listt)}")

def find_min(arr, n):
    '''Iterative'''
    min1 = 0
    for i in range(1, n):
        if arr[i] < arr[i-1]:
            min1 = arr[i]
            
    return min1

print(f"Iterative Minimum Number: {find_min(listt, 5)}")
```

```python
def factorial_recur(n):
    '''Recursive'''
    if n == 1:
        return n
    else:
        return n * factorial_recur(n-1) # recursive
        
length = 15
tic = time.time()
factorial_recur(length)
toc = time.time()
elapsed = toc - tic
print(f"Recursive Factorial: {factorial_recur(length)}, Elapsed: {elapsed}")        
        
def factorial_iter(n):
    '''Iterative'''
    fac = 1
    for i in range(1, n+1):
        fac *= i  # iterative
    return fac

tic = time.time()
factorial_iter(length)
toc = time.time()
elapsed = toc - tic
print(f"Iterative Factorial: {factorial_iter(15)}, Elapsed: {elapsed}")
    
```
