```python
class StackNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        
    def display_node(self):
        print(self.data, end=" -> ")
        
class Stack:
    def __init__(self, data1):
        self.top = StackNode(data1)
        
    def display(self):
        if self.top is None:
            print("Stack is empty")
            return
        current = self.top
        while current:
            current.display_node()
            current = current.next
        print("None")
        
    def push(self, data):
        node = StackNode(data)
        ptr = self.top
        while ptr and ptr.next:
            ptr = ptr.next
        if ptr:
            ptr.next = node
        else:
            self.top = node
        
    def pop(self):
        if self.top is None:
            print("Stack is empty")
            return
        ptr = self.top
        if self.top.next is None:
            temp1 = ptr.data
            self.top = None
            return temp1
        while ptr.next and ptr.next.next:
            ptr = ptr.next
        temp = ptr.next
        ptr.next = None
        return temp.data
    
    def peek(self):
        if self.top is None:
            print("Stack is empty")
            return
        ptr = self.top
        while ptr and ptr.next:
            ptr = ptr.next
        if ptr:
            return ptr.data
        return None
    
    def is_underflow(self):
        return self.top is None
    
    def is_overflow(self):
        try:
            node = StackNode(0)
            return False
        except MemoryError:
            return True
```

**Q**: Find the middle of a linked list without counting the number of nodes of the list.

```python
# Find the middle of a linked list without counting the number of nodes of the list.
'''
1. Initialize both pointers at the head of the linked list.
2. Move fast by two nodes and slow by one node at each iteration.
3. Stop when:
    - fast reaches the end (fast == None), or
    - fast.next == None (for even-length lists).
4. The slow pointer now points to the middle node.
'''
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def find_middle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    return slow.data  # middle node's data


# Example usage:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

print("Middle node:", find_middle(head))
```

**Q**: Swap the first half with the second half of a linked list

```python
# Swap the first half with the second half of a linked list
'''
Approach:
1. Find the middle of the linked list using the slow–fast pointer method.
2. Split the list into two halves.
3. Connect the second half to the first half.
4. Make sure the last node of the list points to None.
'''
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
def swap_halves(head):
    if not head or not head.next:
        return head
    # 1. Find the middle node
    slow = head
    fast = head
    next = None
    while fast and fast.next:
        prev = slow
        slow = slow.next
        fast = fast.next.next
    # 2. Split into halves
    prev.next = None # break link between halves
    second_half = slow
    first_half = head
    # 3. Find end of second half
    temp = second_half
    while temp.next:
        temp = temp.next
    # 4. Attach first half at the end of second half
    temp.next = second_half
    
    return second_half

# Create a linked list: 1 → 2 → 3 → 4 → 5 → 6
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)

# Swap halves
head = swap_halves(head)

# Print swapped list
temp = head
while temp:
    print(temp.data, end=" → " if temp.next else "\n")
```

**Q**: Check if a linked list contains duplicates

```python
# Check if a linked list contains duplicates
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
def contains_duplicates(first):
    seen = set()
    ptr = first
    if ptr:
        if ptr.data in seen:
            return True
        seen.add(ptr.data)
        ptr = ptr.next
    
    return False, seen

# Example list: 1 → 2 → 3 → 2 → 4
first = Node(1)
first.next = Node(2)
first.next.next = Node(3)
first.next.next.next = Node(2)
first.next.next.next.next = Node(4)
first.next.next.next.next.next = Node(3)

print("Contains duplicates?", contains_duplicates(first))
```

**Q**: Remove duplicates from a linked list

```python
# Remove duplicates from a linked list
'''
Algorithm:
1. Use a set to store unique values.
2. Traverse the list.
3. If a node’s value is already in the set → remove it by skipping that node.
4. Else → add the value to the set.
'''
def remove_duplicates(head):
    if not head:
        return None
    seen = set()
    ptr = head
    prev = None
    while ptr:
        if ptr.data in seen:
            # duplicate found, skip it
            prev.next = ptr.next
        else:
            seen.add(ptr.data)
            prev = ptr
        ptr = ptr.next
        
    return head

# Original list: 1 → 2 → 3 → 2 → 4 → 3
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(2)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(3)

print("Before removing duplicates:")
temp = head
while temp:
    print(temp.data, end=" → " if temp.next else "\n")
    temp = temp.next

# Remove duplicates
head = remove_duplicates(head)

print("After removing duplicates:")
temp = head
while temp:
    print(temp.data, end=" → " if temp.next else "\n")
    temp = temp.next
```
