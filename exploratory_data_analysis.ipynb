## 1. **Data Cleaning** ##

### **Task A**: Filter countries of interest (EA countries were chosen)

```python
covid_df = covid_df.rename(columns={'Country/Region': 'Country'})

# filter countries
covid_df = covid_df[covid_df['Country'] \
    .isin(['Kenya', 'Uganda', 'Tanzania', 'Rwanda', 'Burundi'])]
covid_df['Country'].unique()
```

```python
# Check for missing values
msg.matrix(covid_df, figsize=(12, 6))
plt.show()
```
